```python
import pandas as pd
sites = pd.read_excel('sites_info.xlsx')
visitors = pd.read_excel('visitors_with_holidays.xlsx')
```

check if which site do not appear in both tables

```python
a = sites.site_name.unique()
b = visitors.Site_Name.unique()
for x in b:
    if x not in a:
        print(x)
```

```python
visitors.drop(visitors[visitors.Site_Name == 'Afek'].index, inplace=True)
visitors.drop(visitors[visitors.Site_Name == 'Tel Dor'].index, inplace=True)
```

```python
sites.drop('Unnamed: 0', axis=1,inplace=True)
visitors.drop('Unnamed: 0', axis=1,inplace=True)
visitors.drop('Unnamed: 0.1', axis=1,inplace=True)
```
