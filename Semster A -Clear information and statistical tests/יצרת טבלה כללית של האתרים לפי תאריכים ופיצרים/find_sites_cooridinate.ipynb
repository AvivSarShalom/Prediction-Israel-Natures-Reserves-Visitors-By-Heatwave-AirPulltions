```python
import pandas as pd
```

```python
df_site = pd.read_excel('sites_info_with_dummies.xlsx')
df_site.drop('Unnamed: 0',axis=1, inplace=True)
sites=df_site.site_name
sites
```

```python
from geopy.geocoders import Nominatim
loc = Nominatim(user_agent="GetLoc")
coo = {}
for city in sites:
    try:
        getLoc = loc.geocode(city)
        if(getLoc is not type(None)):
            coo[city] = [getLoc.latitude,getLoc.longitude]
    except :
        continue
```

```python
len(coo)
len(sites)
```

```python
df=pd.DataFrame.from_dict(coo)
df=df.transpose()
df.columns=['lat','lon']



import folium

lat = df["lat"][0]
lon = df["lon"][0]

map = folium.Map(location=[lat, lon], zoom_start=3)
for i in range (len(df.lat)):
    folium.Marker([df["lat"][i], df['lon'][i]],popup= folium.Popup(df.index[i])).add_to(map)
    folium.Circle([df["lat"][i], df['lon'][i]]
    # ,radius=5000
                       ).add_to(map)

map
```
