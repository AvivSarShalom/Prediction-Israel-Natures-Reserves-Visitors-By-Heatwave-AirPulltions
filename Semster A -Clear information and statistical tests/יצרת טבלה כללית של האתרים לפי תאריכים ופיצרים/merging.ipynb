```python
import pandas as pd
sites = pd.read_excel('sites_info_with_dummies.xlsx')
visitors = pd.read_excel('visitors_with_holidays.xlsx')
```

```python
sites.drop('Unnamed: 0', axis=1, inplace=True)
visitors.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
sites.rename({'site_name':'Site_Name'},inplace=True, axis=1)
```

```python
sites.head()
```

```python
df = visitors.merge(right = sites, on='Site_Name')
```

```python
df.sort_values(by='Date',inplace=True)
df.reset_index(drop=True,inplace=True)
df.head()
```

```python
df.to_excel('test.xlsx')
```
