```python
muslims_holidays = pd.read_excel('muslims_holidays.xlsx')
muslims_holidays_list = list(muslims_holidays['dates'])
```

```python
muslims_holidays_dates = []
```

```python
#add muslim holidays 

dates = [str(x).split()[0] for x in muslims_holidays_list]
for date in dates:
    muslims_holidays_dates.append(date)
```

get all the vacations for muslims

```python
winter_break = []
```

```python
years = [2016,2017,2018,2019,2020,2021] #add all the winter breaks
for i in range(len(years)):
    tmp = list(pd.to_datetime(pd.date_range(start=f"{years[i]}-12-24",end=f"{years[i]+1}-01-08")))
    dates = [str(x).split()[0] for x in tmp]
    for date in dates:
        winter_break.append(date)

winter_break
```

```python
#muslims spring break
tmp1 = list(pd.to_datetime(pd.date_range(start="2016-03-23",end="2016-04-11")))
tmp2 = list(pd.to_datetime(pd.date_range(start="2017-04-04",end="2017-04-18")))
tmp3 = list(pd.to_datetime(pd.date_range(start="2018-03-29",end="2018-04-13")))
tmp4 = list(pd.to_datetime(pd.date_range(start="2019-04-15",end="2019-05-01")))
tmp5 = list(pd.to_datetime(pd.date_range(start="2020-04-06",end="2020-04-21")))
tmp6 = list(pd.to_datetime(pd.date_range(start="2021-03-30",end="2021-04-17")))

spring_break = [tmp1,tmp2,tmp3,tmp4,tmp5,tmp6]
spring_break_dates = []
for x in spring_break:
    dates = [str(i).split()[0] for i in x]
    for date in dates:
        spring_break_dates.append(date)

spring_break_dates
```

```python
summer_break = []
```

```python
for i in range(len(years)):
    tmp = list(pd.to_datetime(pd.date_range(start=f"{years[i]}-06-20",end=f"{years[i]}-09-01")))
    dates = [str(x).split()[0] for x in tmp]
    for date in dates:
        summer_break.append(date)
```

```python
summer_break
```

```python
jewish_vacations = summer_break +elections
muslims_vacations = spring_break_dates+winter_break+summer_break+elections
```

```python
df['is_jewish_holiday'] = [1 if str(val).split()[0] in israel_holidays else 0 for val in df['Date']]
df['is_jewish_vacation']= [1 if str(val).split()[0] in jewish_vacations else 0 for val in df['Date']]
df['is_muslims_holiday']= [1 if str(val).split()[0] in muslims_holidays_dates else 0 for val in df['Date']]
df['is_muslims_vacation']= [1 if str(val).split()[0] in muslims_vacations else 0 for val in df['Date']]
```
