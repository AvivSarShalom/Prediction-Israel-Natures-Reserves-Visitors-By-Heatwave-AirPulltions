```python
#determine whether a date is on weekend or not
def is_weekend(date):
    res = datetime.weekday(date)
    if res==4 or res==5:
        return True

    return False
```

```python
from datetime import datetime
#datetime.weekday()
df['is_weekend'] = df['Date'].apply(lambda x: 1 if is_weekend(x) else 0)
```
