```python
import pandas as pd
```

```python
df = pd.read_excel("complete_site_info.xlsx")
print(df.columns)
df.head()
```

```python
#setting all the default parameters like empty columns, thresholds etc.
pm10_threshold = 130
pm25_threshold = 37.5
nox_threshold = 25
so2_threshold = 50
df['is_Site_exceeded_pm10']=0
df['is_Site_exceeded_pm2.5']=0
df['is_Site_exceeded_nox']=0
df['is_Site_exceeded_so2']=0
df['is_Site_exceeded_pm10'][df['pm10']>pm10_threshold]=1
df['is_Site_exceeded_pm2.5'][df['pm2.5']>pm25_threshold]=1
df['is_Site_exceeded_so2'][df['so2']>=nox_threshold]=1
df['is_Site_exceeded_nox'][df['nox']>=so2_threshold]=1

print("pm10 |", df["Date"][df['is_Site_exceeded_pm10']==1].count(),"|",df["Date"][df['is_Site_exceeded_pm10']==1].count()/len(df))
print("pm2.5|",df["Date"][df['is_Site_exceeded_pm2.5']==1].count(),"|",df["Date"][df['is_Site_exceeded_pm2.5']==1].count()/len(df))
print("nox|",  df["Date"][df['is_Site_exceeded_nox']==1].count(),"|",df["Date"][df['is_Site_exceeded_nox']==1].count()/len(df))
print("so2|" , df["Date"][df['is_Site_exceeded_so2']==1].count(),"|",df["Date"][df['is_Site_exceeded_so2']==1].count()/len(df))
print("------------------------------------------------------------")
```

```python
pm10_city_list=['Tel_Aviv-Yafo_pm10', 'Jerusalem_pm10', 'Haifa_pm10', 'Ashkelon_pm10','Beer-Sheva_pm10', 'Eilat_pm10', 'Qiryat Shemona_pm10', 'Safed_pm10']
pm25_city_list=['Tel_Aviv-Yafo_pm2.5', 'Jerusalem_pm2.5', 'Haifa_pm2.5','Ashkelon_pm2.5', 'Beer-Sheva_pm2.5', 'Eilat_pm2.5','Qiryat Shemona_pm2.5', 'Safed_pm2.5']
nox_city_list=['Tel_Aviv-Yafo_nox','Jerusalem_nox', 'Haifa_nox', 'Ashkelon_nox', 'Beer-Sheva_nox','Eilat_nox', 'Qiryat Shemona_nox', 'Safed_nox']
so2_city_list=[ 'Tel_Aviv-Yafo_so2','Jerusalem_so2', 'Haifa_so2', 'Ashkelon_so2', 'Beer-Sheva_so2','Eilat_so2', 'Qiryat Shemona_so2', 'Safed_so2']
```

```python
(((df[pm10_city_list]>pm10_threshold).sum(axis=1))>=1)
(((df[pm25_city_list]>pm25_threshold).sum(axis=1))>=1)
(((df[so2_city_list]>so2_threshold).sum(axis=1)  )>=1)
(((df[nox_city_list]>nox_threshold).sum(axis=1)  )>=1)
```

```python
#if a pollutant exceeded 
def f1(type_threshold,listType):
    for c in listType:
        a = [ 1 if x else 0 for x in (df[c]>type_threshold) ]
        name = c+'_exceeded'
        df.insert(len(df.columns),name,a)

```

```python
pd.set_option("display.max_columns", None)
df.head()
```

```python
f1(pm10_threshold,pm10_city_list)
f1(pm25_threshold,pm25_city_list)
f1(so2_threshold,so2_city_list)
f1(nox_threshold,nox_city_list)
df.head()
```
