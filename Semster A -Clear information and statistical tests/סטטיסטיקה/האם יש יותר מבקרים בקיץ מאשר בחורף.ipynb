```python
import pandas as pd
import numpy as np
import datetime 
```

```python
df = pd.read_excel("complete_site_info.xlsx")
```

```python
df.head()
```

```python
summer = df[['Date',"Total"]][
    ((df["Date"]>=pd.to_datetime('2016-6-1')) & (df["Date"]<pd.to_datetime('2016-9-1')))|
    ((df["Date"]>=pd.to_datetime('2017-6-1')) & (df["Date"]<pd.to_datetime('2017-9-1')))|
    ((df["Date"]>=pd.to_datetime('2018-6-1')) & (df["Date"]<pd.to_datetime('2018-9-1')))|
    ((df["Date"]>=pd.to_datetime('2019-6-1')) & (df["Date"]<pd.to_datetime('2019-9-1')))|
    ((df["Date"]>=pd.to_datetime('2020-6-1')) & (df["Date"]<pd.to_datetime('2020-9-1')))|
    ((df["Date"]>=pd.to_datetime('2021-6-1')) & (df["Date"]<pd.to_datetime('2021-9-1')))]
summer

winter =  df[['Date',"Total"]][
    (df["Date"] < pd.to_datetime('2016-3-1')) | 
    ((df["Date"]>=pd.to_datetime('2016-12-1')) & (df["Date"]<pd.to_datetime('2017-3-1')))|
    ((df["Date"]>=pd.to_datetime('2017-12-1')) & (df["Date"]<pd.to_datetime('2018-3-1')))|
    ((df["Date"]>=pd.to_datetime('2018-12-1')) & (df["Date"]<pd.to_datetime('2019-3-1')))|
    ((df["Date"]>=pd.to_datetime('2019-12-1')) & (df["Date"]<pd.to_datetime('2020-3-1')))|
    ((df["Date"]>=pd.to_datetime('2020-12-1')) & (df["Date"]<pd.to_datetime('2021-3-1')))]
winter

spring = df[['Date',"Total"]][
    ((df["Date"]>=pd.to_datetime('2016-3-1')) & (df["Date"]<pd.to_datetime('2016-6-1')))|
    ((df["Date"]>=pd.to_datetime('2017-3-1')) & (df["Date"]<pd.to_datetime('2017-6-1')))|
    ((df["Date"]>=pd.to_datetime('2018-3-1')) & (df["Date"]<pd.to_datetime('2018-6-1')))|
    ((df["Date"]>=pd.to_datetime('2019-3-1')) & (df["Date"]<pd.to_datetime('2019-6-1')))|
    ((df["Date"]>=pd.to_datetime('2020-3-1')) & (df["Date"]<pd.to_datetime('2020-6-1')))|
    ((df["Date"]>=pd.to_datetime('2021-3-1')) & (df["Date"]<pd.to_datetime('2021-6-1')))]
spring

autumn = df[['Date',"Total"]][
    ((df["Date"]>=pd.to_datetime('2016-9-1')) & (df["Date"]<pd.to_datetime('2016-12-1')))|
    ((df["Date"]>=pd.to_datetime('2017-9-1')) & (df["Date"]<pd.to_datetime('2017-12-1')))|
    ((df["Date"]>=pd.to_datetime('2018-9-1')) & (df["Date"]<pd.to_datetime('2018-12-1')))|
    ((df["Date"]>=pd.to_datetime('2019-9-1')) & (df["Date"]<pd.to_datetime('2019-12-1')))|
    ((df["Date"]>=pd.to_datetime('2020-9-1')) & (df["Date"]<pd.to_datetime('2020-12-1')))|
    ((df["Date"]>=pd.to_datetime('2021-9-1')) & (df["Date"]<pd.to_datetime('2021-12-1')))]
autumn
```
