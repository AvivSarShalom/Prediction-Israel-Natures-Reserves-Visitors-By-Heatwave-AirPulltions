```python
import pandas as pd
df = pd.read_excel("tmp.xlsx")

df.head(50)
```

```python
import datetime
for i in range (len(df)):
    x=df.Date[i].split('-')
    df.Date[i] = datetime.datetime(int(x[2]), int(x[1]), int(x[0]))
```

```python
df["Date"].head(50)
```

```python
df["Date"] = pd.to_datetime( df["Date"])
df["Date"].head(50)
```

```python

for c in df.columns:
    if "מאוישת" in c:
        df.drop(c,axis=1,inplace=True)

df.head()
```

```python
df = pd.melt(df, 'Date')
df = df.sort_values("Date")
df.reset_index(drop=True,inplace=True)
df.columns = ["Date","City","Temperature"]
```

```python
import numpy as np
df['Temperature'] = df['Temperature'].replace("-",np.NaN)
df['Temperature'] = df['Temperature'].astype(np.float64)
```

```python
df[df["Temperature"]>40].head(50)
```

```python
cities = df.City.unique()
cities=sorted(cities)
cities
# cities_new = [ city.split('') for city in cities  ]
# cities_new

```

```python
from geopy.geocoders import Nominatim
loc = Nominatim(user_agent="GetLoc")
coo = {}
for city in cities:
    try:
        getLoc = loc.geocode(city)
        if(getLoc is not type(None)):
            coo[city] = [getLoc.latitude,getLoc.longitude]
    except :
        continue



print(len(cities))
len(coo)
for c in cities:
    if c not in coo:
        print(c)

df['latitude'] = 0
df['longitude'] = 0


map
```
