```python
import pandas as pd
```

```python
pm25_df = pd.read_excel("pm25_T_NEW_WITH_NO_DROP_CITIES.xlsx",date_parser={"Date"})
pm10_df = pd.read_excel("pm10_T_NEW_WITH_NO_DROP_CITIES.xlsx",date_parser={"Date"})
Nox_df = pd.read_excel("Nox_T_NEW_WITH_NO_DROP_CITIES.xlsx",date_parser={"Date"})
So2_df = pd.read_excel("So2_T_NEW_WITH_NO_DROP_CITIES.xlsx",date_parser={"Date"})
So2_df.info()
```

```python
So2_df.head()
```

```python
merage_df = pm10_df.merge(right=pm25_df,on=["Date","City"],how="outer")
merage_df = merage_df.merge(right=Nox_df,on=["Date","City"],how="outer")
merage_df = merage_df.merge(right=So2_df,on=["Date","City"],how="outer")
merage_df
```

```python
merage_df.columns=['Date',"City",'pm10','pm2.5','nox','so2']
merage_df.sort_values(["Date"],inplace=True)
merage_df
```

```python
print(len(merage_df))
merage_df=merage_df.dropna(thresh=3)
cities = merage_df.City.unique()
len(merage_df)
```

```python
print(cities)
```

```python
from geopy.geocoders import Nominatim
loc = Nominatim(user_agent="GetLoc")
coo = {}
for city in cities:
    try:
        getLoc = loc.geocode(city)
        if(getLoc is not type(None)):
            coo[city] = [getLoc.latitude,getLoc.longitude]
    except :
        continue



```

```python
print(len(cities))
len(coo)
for c in cities:
    if c not in coo:
        print(c)
```

```python
coo['מועצה תעשייתית נאות חובב'] = [31.135,34.791]
coo['קיריית טבעון'] = [32.715603,35.126578]
coo['חיפה וקריות'] = [32.841592,35.071278]
coo['ארז'] =[31.56028,34.566483]
coo['מגל'] =[32.385837,35.03366]
coo['ערד'] =[31.258928,35.213893]
coo['גן דרום'] = [31.804468,34.700824]

coo['ברקאי'] = [32.474819141691206,35.0275950342368]
coo['חריש'] = [32.462194973207886,35.045251620627795]
coo['קיסריה'] = [32.50712927639148,34.91244095275816]
coo['אלעד'] = [32.05089076091389,34.95718705819683]
coo['אשלים'] = [30.963449095763554,34.70081419534035]
```

```python
merage_df['latitude'] = 0
merage_df['longitude'] = 0
```

```python
for key,val in coo.items():
    merage_df.latitude[merage_df["City"]==key] = val[0]
    merage_df.longitude[merage_df["City"]==key] = val[1]

merage_df.reset_index(drop=True,inplace=True)
```

```python
df=pd.DataFrame.from_dict(coo)
df=df.transpose()
df.columns=['lat','lon']

import folium

lat = df["lat"][0]
lon = df["lon"][0]

map = folium.Map(location=[lat, lon], zoom_start=3)
for i in range (len(df.lat)):
    folium.Marker([df["lat"][i], df['lon'][i]],popup= folium.Popup(df.index[i])).add_to(map)
    folium.Circle([df["lat"][i], df['lon'][i]]).add_to(map)

map
```
