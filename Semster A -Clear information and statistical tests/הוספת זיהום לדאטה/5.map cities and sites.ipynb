```python

import math

def get_distance(lat_1, lng_1, lat_2, lng_2): 
    lng_1, lat_1, lng_2, lat_2 = map(math.radians, [lng_1, lat_1, lng_2, lat_2])

    d_lat = lat_2 - lat_1
    d_lng = lng_2 - lng_1 

    temp = (  
         math.sin(d_lat / 2) ** 2 
       + math.cos(lat_1) 
       * math.cos(lat_2) 
       * math.sin(d_lng / 2) ** 2
    )

    return 6373.0 * (2 * math.atan2(math.sqrt(temp), math.sqrt(1 - temp)))
# lat1=32.20089
# lon1=35.2731

# lat2=32.10608
# lon2=35.18514
# get_distance(lat1,lon1,lat2,lon2)

nearst_withot_dst={}
nearst={}
cities = df_cities["Loction"]
latC = df_cities["latitude"]
lonC = df_cities["longitude"]
sites = df_sites['Loction']
latS = df_sites['latitude']
lonS = df_sites['longitude']


for i in range (len(df_sites)):
    print(i)
    tmp={}
    for j in range (len(df_cities)):
        # dst = get_distance(lonC[j] ,latC[j] , lonS[i] ,lonS[i] )
        dst = get_distance(latC[j] ,lonC[j] , latS[i] ,lonS[i] )
        tmp[df_cities.Loction[j]] = dst
    min_list =(sorted(tmp.items(), key=lambda item: item[1]))
    min_list_dict =dict(sorted(tmp.items(), key=lambda item: item[1]))
    nearst_withot_dst[df_sites.Loction[i]]=min_list
    nearst[df_sites.Loction[i]]=min_list_dict



```
