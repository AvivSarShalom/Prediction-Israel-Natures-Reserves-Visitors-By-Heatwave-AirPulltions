```python
types = ["nox","pm10","pm2.5","so2"]
Site_info_type_stay={}
for site in sites:
    Site_info_type_stay[site] =[]
    for t in types:
        per= round(len(df[["Date","Site_Name","pm10"]][(df["Site_Name"]==site) & (df[t].isna())]) / len(df[df["Site_Name"]==site])*100,2)
        print(site," null %% in ",t ,per)
        if per<40.00 : 
            Site_info_type_stay[site].append(t)

    
```

```python
models={}
i=0
for site,types in Site_info_type_stay.items():
    print(site,types)
    if types not in models.values():
        models[i]=types
        i+=1
    # else:
    #     models[types].append(site)

models
```

```python
df.insert(0,"Model_number", -1)
for site in sites:
    df["Model_number"][df["Site_Name"]==site]=Site_info_type_stay[site][-1]

df
```

```python
pivot_df=df.pivot_table(index=["Site_Name","Date"],values=["Model_number","pm10","pm2.5","nox","so2","Temperature"])
pivot_df
```

```python
dfs = []
pivot_df["Model_number"][pivot_df.index.get_level_values("Site_Name")==site][0]
i=0
for site in sites:
    dfs.append( pivot_df[pivot_df.index.get_level_values("Site_Name")==site])
    dfs[i]=dfs[i].reset_index().set_index("Date")
    i=i+1

i=0
for j in range (len(dfs)):
    for t in models[dfs[j]["Model_number"][0]]:
       dfs[j][t] = round(dfs[j][t].interpolate(method="time",limit_direction="both"),2)
    dfs[j]["Temperature"] = round(dfs[j]["Temperature"].interpolate(method="time",limit_direction="both"),2)

for j in range(len(dfs)):
    dfs[j]["is_HeatWave"]=0
    for i in range(2,len(dfs[j])):
        if dfs[j]['Temperature'][i-2]>30 and dfs[j]['Temperature'][i-1]>30 and dfs[j]['Temperature'][i]>30:
            dfs[j]['is_HeatWave'][i-2]=1
            dfs[j]['is_HeatWave'][i-1]=1
            dfs[j]['is_HeatWave'][i]=1


        
```
