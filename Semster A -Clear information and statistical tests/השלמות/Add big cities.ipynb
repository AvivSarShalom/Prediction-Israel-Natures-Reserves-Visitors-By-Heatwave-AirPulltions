```python

import math
from geopy.distance import geodesic 

def get_distance(lat_1, lng_1, lat_2, lng_2): 
    # lng_1, lat_1, lng_2, lat_2 = map(math.radians, [lng_1, lat_1, lng_2, lat_2])

    # d_lat = lat_2 - lat_1
    # d_lng = lng_2 - lng_1 

    # temp = (  
    #      math.sin(d_lat / 2) ** 2 
    #    + math.cos(lat_1) 
    #    * math.cos(lat_2) 
    #    * math.sin(d_lng / 2) ** 2
    # )

    # return 6373.0 * (2 * math.atan2(math.sqrt(temp), math.sqrt(1 - temp)))
    # return math.sqrt((lat_1-lat_2)**2+(lng_1-lat_2)**2)
    coords_1 = (lat_1,lng_1)
    coords_2 = (lat_2,lng_2)
    return geodesic (coords_1, coords_2).km


nearst={}
cities = df_dict["index"]
latC = df_dict["lat"]
lonC = df_dict["lon"]
sites = cities_choose


for i in range (len(sites)):
    print(i)
    tmp={}
    for j in range (len(cities)):
        # dst = get_distance(lonC[j] ,latC[j] , lonS[i] ,lonS[i] )
        dst = get_distance(latC[j] ,lonC[j] , coordinates[i][0] ,coordinates[i][1] )
        tmp[cities[j]] = dst
    min_list_dict =dict(sorted(tmp.items(), key=lambda item: item[1]))
    nearst[sites[i]]=min_list_dict
nearst
```
