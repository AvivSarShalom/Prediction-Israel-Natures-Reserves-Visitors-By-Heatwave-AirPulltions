```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
df = pd.read_excel("../site_info_ver_3.3.xlsx")
```

```python
tmp = df
index =  tmp.loc[tmp.Site_Name=='Prat Stream-En Mabo‘a'].index
tmp.drop(index,inplace=True)
index =  tmp.loc[tmp.Site_Name=='Baptismal Site Qasr al-Yahud'].index
tmp.drop(index,inplace=True)
tmp["Day_before_Total"] = np.nan
```

```python
sites = tmp.Site_Name.unique()
```

```python

diff_day = pd.Timedelta('1 days 00:00:00')
for site in sites:
    site_df = tmp.loc[tmp.Site_Name==site]
    for i in range(1,len(site_df.index)):
        if site_df.iloc[i,0] - site_df.iloc[i-1,0]==diff_day:
            site_df.iloc[i,-1] =site_df.iloc[i-1,4] 

    if 1-len(site_df.loc[site_df.Day_before_Total>=0,"Day_before_Total"])/len(site_df)<0.4:
        site_df.Day_before_Total.interpolate(method='nearest',limit_direction="both",inplace=True)
        site_df.Day_before_Total.fillna(site_df.Day_before_Total.median(),inplace=True)
    else : print(site+" Not filled nan value",1-len(site_df.loc[site_df.Day_before_Total>=0,"Day_before_Total"])/len(site_df))
    site_df=site_df[['Day_before_Total']]
    tmp.loc[tmp.Site_Name==site,'Day_before_Total']=site_df.Day_before_Total
    # tmp = tmp.merge(right=site_df,left_index=True,right_index=True,how='')
```

```python
tmp.Day_before_Total = tmp.Day_before_Total.astype('int64')
```
